### Marking outliers on the DataFrame

```python
charges = df.charges

q75, q25 = np.percentile(charges, [75 ,25])
iqr = q75 - q25

cut_off = iqr * 1.5
cut_off_lower = q25 - cut_off
cut_off_upper = q75 + cut_off

df['outlier_or_not'] = ['lower' if i < cut_off_lower else 'upper' if i > cut_off_upper else 'nothing' for i in charges]
```

### Entries with cheapest insurance charges

```python
at_min = df[df.charges==df.charges.min()]
at_min
```

- nothing in the outlier_or_not column means that the charge is not an outlier

```python
min_age = df.age.min()

print('Youngest candidate is of age:',min_age,'years')
```

- The cheapest insurance charge is 1121.87, paied by the youngest male of age 18 years, who is a none smoker, having no children , with a bmi of 23.21 living in the southwest region

### Entries with highest insurance charges

```python
at_max = df[df.charges==df.charges.max()]
at_max
```

- upper in the outlier_or_not column means that the charge is an outlier in the upper percentile

### Observing upper ouliers 

```python
upper_oulier = df[df['outlier_or_not'] == 'upper']
up = upper_oulier[['sex','smoker','age','bmi','children', 'charges']]
upper_oulier.tail()
```

### How different genders are charged 

```python
sns.kdeplot(df[df['sex']=='female']['charges'], shade=True, label = 'Female charge')
sns.kdeplot(df[df['sex']=='male']['charges'], shade=True, label = 'Male charge')
plt.title('Gender as per charges')
plt.xlabel('charges')
plt.ylabel('count');
```

- Females are paying more charges than males


### Relation of Gender and age

```python
sns.kdeplot(df[df['sex']=='female']['age'], shade=True, label = 'Female age')
sns.kdeplot(df[df['sex']=='male']['age'], shade=True, label = 'Male age')
plt.title('Age pe gender')
plt.xlabel('age(years)')
plt.ylabel('count');
```

- There are more males between the ages of 18 years and 40 years compared to females
- There are more females in ages between 40 years and 65 years compared to males

```python
sns.kdeplot(df[df['sex']=='female']['bmi'], shade=True, label = 'Female bmi')
sns.kdeplot(df[df['sex']=='male']['bmi'], shade=True, label = 'Male bmi')
plt.title('Gender vs BMI')
plt.xlabel('bmi')
plt.ylabel('count');
```

- males are found to be with the highest bmi

```python
sns.swarmplot(x='sex', y='charges', data =df)
```

```python
#impact of smoke on charges

df.groupby('smoker')['charges'].agg('mean').plot.bar()
```

```python
#Plotting Scatter diagram for bmi vs charges

data = [go.Scatter(x=df.bmi, y=df.charges,mode ='markers',
                  marker = dict(color='rgb(51,120,3)'))]
layout = go.Layout(title = 'Influence of age on insurance charges', 
                  xaxis=dict(title='bmi'), yaxis=dict(title='charges'),
                  hovermode = 'closest')

fig = go.Figure(data=data, layout =layout)
pyo.plot(data,filename ='scatter.html')
```
